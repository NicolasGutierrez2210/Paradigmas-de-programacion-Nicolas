```python
import time
import matplotlib.pyplot as plt

!pip install memory-profiler

def facto_r(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * facto_r(n - 1)


def facto_i(n):
    resultado = 1
    for i in range(2, n + 1):
        resultado *= i
    return resultado



n = 200  # Lo podemos modificar a disposicion


start_time = time.time()
facto_r(n)
end_time = time.time()
print(f"Tiempo de ejecución (recursivo) para {n}!: {end_time - start_time:.10f} segundos")


start_time = time.time()
facto_i(n)
end_time = time.time()
print(f"Tiempo de ejecución (iterativo) para {n}!: {end_time - start_time:.10f} segundos")




from memory_profiler import memory_usage

# memoria para recursivo
memoria_r = memory_usage((facto_r, (n,)))
# memoria para iterativo
memoria_i = memory_usage((facto_i, (n,)))

print(f"Uso de memoria para recursivo (recursivo): {max(memoria_r)} MiB")
print(f"Uso de memoria para iterativo (iterativo): {max(memoria_i)} MiB")





import matplotlib.pyplot as plt

n_values = list(range(1, 200))  # Se modifica el rango
tiempos_r = []
tiempos_i = []

for n in n_values:
    # Tiempo de ejecucion recursivo
    start = time.time()
    facto_r(n)
    tiempos_r.append(time.time() - start)

    # Tiempo de ejecucion iterativo
    start = time.time()
    facto_i(n)
    tiempos_i.append(time.time() - start)

# Graficas
plt.figure(figsize=(10,6))
plt.plot(n_values, tiempos_r, label='Recursivo')
plt.plot(n_values, tiempos_i, label='Iterativo')
plt.xlabel('Valor de n')
plt.ylabel('Tiempo de ejecucion (s)')
plt.title('Comparacion entre el Tiempo de Ejecucion: Recursivo vs Iterativo')
plt.legend()
plt.grid(True)
plt.show()
```
